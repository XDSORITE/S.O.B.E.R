```python
import os
import yaml

base = os.path.abspath("data")

data_yaml_content = {
    "path": base,
    "train": os.path.join(base, "train/images"),
    "val":   os.path.join(base, "valid/images"),
    "test":  os.path.join(base, "test/images"),
    "nc": 1,
    "names": ["seatbelt"]
}

with open("data/data.yaml", "w") as f:
    yaml.dump(data_yaml_content, f, default_flow_style=False)

print(open("data/data.yaml").read())
```

```python
train_py = '''import os
from ultralytics import YOLO

def main():
    model = YOLO("yolo11n.pt")
    model.train(
        data="data/data.yaml",
        epochs=50,
        imgsz=640,
        device="0,1",
        batch=16,
        project="runs/train",
        name="seatbelt",
        exist_ok=True,
    )
    print("done - best weights at runs/train/seatbelt/weights/best.pt")

if __name__ == "__main__":
    main()
'''

with open("train.py", "w") as f:
    f.write(train_py)

print("wrote train.py")
```
